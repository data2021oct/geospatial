```python
import folium
from folium import Choropleth, Circle, Marker, Icon, Map, TileLayer
from folium.plugins import HeatMap, MarkerCluster
import geopandas as gpd
import pandas as pd
```

```python
 localizations = {'Austin': ['30.27125853189173', '-97.75551842126049'],
 'Hamburg': ['53.55687309076416', '10.002900693677232'],
 'Newcastle': ['54.96923892979405', '-1.6162802662146134']}
```

```python
venues = pd.read_csv("../data/venues_df.csv")
```

```python
Austin_df = venues[venues["city"] == "Austin"].sort_values(["category","distance"])
Austin_df.head(3)
```

```python
Austin_df.category.unique()
```

```python
aus_lat = localizations["Austin"][0]
aus_lon = localizations["Austin"][1]
```

```python
map_aus = folium.Map(location=[aus_lat,aus_lon], zoom_start=15)
```

```python
for i,row in Austin_df.iterrows():
    cat = {"location": [row["latitud"], row["longitud"]], "tooltip": row["category"]}
    
    if row["category"] == "Nursery School" or row["category"] == "Elementary School" or row["category"] == "Preschool" :
        icono = Icon(color = "red",
                     prefix="fa",
                     icon="child",
                     icon_color="black" )
        mark = Marker(**cat, icon=icono)
        mark.add_to(map_aus)
        
    elif row["category"] == "Coffee Shop":
        icono = Icon(color = "beige",
                     prefix="fa",
                     icon="coffee",
                     icon_color="black")
        mark = Marker(**cat, icon=icono)
        mark.add_to(map_aus)
        
    elif row["category"] == "Pet Service" or row["category"] == "Pet Store" or row["category"] == "Veterinarian" :
        icono = Icon(color = "cadetblue",
                     prefix="fa",
                     icon="paw",
                     icon_color="black")
        
        mark = Marker(**cat, icon=icono)
        mark.add_to(map_aus)
```
